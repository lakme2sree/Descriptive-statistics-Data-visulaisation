# dealing with missing values

```python
cardf.describe()
```

## here hp is missing. means there is something wrong with the hp column, could be missing values

```python
cardf.info()
```

```python
cardf.dtypes
```

```python
ishpdigit= pd.DataFrame(cardf.hp.str.isdigit())
```

```python
ishpdigit
```

```python
cardf[ishpdigit['hp']==False]
```

```python
cardf=cardf.replace('?',np.nan)
```

```python
cardf[ishpdigit['hp']==False]
```

```python
medianfiller = lambda x : x.fillna(x.median())
cardf=cardf.apply(medianfiller,axis=0)    
```

## Every column's missing value is replaced with that columns median value respectively. No need to specify the column . Row wise the median value is being added to the dataframe.

```python
cardf['hp']=cardf['hp'].astype('float64')
```

### converting the dtype of hp to float. It was object before

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```
